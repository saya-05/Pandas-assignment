```python
import pandas as pd
```

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',sep='\t')
```

```python
chipo
```

```python
#How many products cost more than $10.00?
dollar = lambda x:float(x[1:-1])
chipo.item_price = chipo.item_price.apply(dollar)
```

```python
chipo['item_price'].dtypes
```

```python
chipo[chipo['item_price']>10].count()
```

```python
#What is the price of each item?
chipo['price_of_each_quantity'] = chipo['item_price']/chipo['quantity']
```

```python
chipo
```

```python
result = chipo[['item_name','price_of_each_quantity']]
```

```python
result.sort_values(by=['item_name','price_of_each_quantity'],ascending = True).head(100)
```

```python
result.drop_duplicates(subset=['item_name'])
```
